```python
import math

def circleAreaCoverage(radiusOfCircle1, radiusOfCircle2):
    if not isinstance(radiusOfCircle1, int) or not isinstance(radiusOfCircle2, int):
        return "Invalid input: radii must be positive integers."

    if radiusOfCircle1 <= 0 or radiusOfCircle2 <= 0:
        return "Invalid input, the input must be positive."

    areaOfCircle1 = math.pi * (radiusOfCircle1**2)
    areaOfCircle2 = math.pi * (radiusOfCircle2**2)

    larger_area = max(areaOfCircle1, areaOfCircle2)
    smaller_area = min(areaOfCircle1, areaOfCircle2)

    areaCoveragePercentage = (smaller_area / larger_area) * 100
    return f"{areaCoveragePercentage:.2f} %"

print(circleAreaCoverage(-2, 5))
print()
print(circleAreaCoverage(3, 5))
```
